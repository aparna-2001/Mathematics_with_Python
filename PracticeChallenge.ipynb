```python
#1. Even odd vending machine



num = float(input("Enter a number: "))

if num.is_integer():
    num = int(num)
    print("You have entered an integer")

    if num % 2 == 0:
        print("even")
        print(num)
        for i in range(1, 10):
            print(num + (i * 2))
    else:
        print("odd")
        print(num)
        for i in range(1, 10):
            print(num + (i * 2))

else:
    print("Enter a valid integer")
```

```python
#3. Enhanced unit converter


def print_menu():
    print("1. kg to pounds")
    print("2. pounds to kg")
    print("3. fahrenheit to celsius")
    print("4. Celsius to fahrenheit")
    print("5. Miles to km")
    print("6. Km to miles")


def kg_pounds():
    kg = float(input ("Enter the mass in kg: "))
    pounds = kg * 2.205

    print("mass in pounds {0}".format(pounds))

def pounds_kg():
    pounds = float(input ("Enter the mass in pounds: "))
    kg = pounds / 2.205

    print("mass in kg {0}".format(kg))


def f_celsius():
    f = float(input ("Enter the temperature in fahrenheit: "))
    celsius = (f-32)*(5/9)

    print("Temperature in celsius {0}".format(celsius))

def celsius_f():
    celsius = float(input ("Enter the mass in kg: "))
    f = (9/5) * celsius + 32

    print("Temperature in fahrenheit {0}".format(f))

def miles_km():
    miles = float(input("Enter distance in miles: "))
    km = miles * 1.609

    print("Distance in miles: {0}".format(km))


def km_miles():
    km = float(input("Enter distance in km: "))
    miles = km / 1.609
    print("Distance in km: {0}".format(miles))


print_menu()
choice = input("Which conversion would you like to do: ")

if choice == '1':
    kg_pounds()
elif choice == '2':
    pounds_kg()
elif choice == '3':
    f_celsius()
elif choice == '4':
    celsius_f()
elif choice == '5':
    miles_km()
elif choice == '6':
    km_miles()
else:
    print("Invalid choice")
```

```python

#4. Fraction calculator



from fractions import Fraction

def print_operations():
    print('add')
    print('subtract')
    print('multiply')
    print('divide')



def add(a,b):
    print("Result of addition: {0}".format(a+b))

def subtract(a,b):
    print("Result of subtraction: {0}".format(a-b))

def multiply(a,b):
    print("Result of multiplication {0}".format(a*b))

def divide(a,b):
    print("Result of division: {0}".format(a/b))


a = Fraction(input("Enter first fraction: "))
b = Fraction(input ("Enter second fraction: "))

print_operations()
choice = input("Enter your choice of operation: ")

if choice=="add":
    add(a,b)
elif choice=="subtract":
    subtract(a,b)
elif choice=="multiply":
    multiply(a,b)
elif choice=="divide":
    divide(a,b)
else:
    print("invalid choice")
```

```python
#Exploring Quadratic function visually


from matplotlib import pyplot as plt
import math
from pylab import axis, savefig


def draw_graph(x_values,y_list):
   plt.xlabel('x values')
   plt.ylabel('y values')
   plt.title('visualising quadratic function')
   axis(xmin = -5,xmax=5, ymin= 0, ymax=36)
   plt.plot(x_values,y_list, marker='o')



y_list = []


x_values = [-5,-4,-3,-2,-1,1,2,3,4,5]
for x in x_values:
     y = x**2 + 2*x + 1
     print("(x={0}, y={1})".format(x,y))
     y_list.append(y)

draw_graph(x_values,y_list)
plt.show()
savefig('quadratic_graph.png')
```

```python

#Relationship between the Fibonacci sequence and the golden ratio



from matplotlib import pyplot as plt
import math

def fibo(n):
    if n == 1:
        return [1]
    if n == 2:
        return [1,1]

    a = 1
    b = 1
    series = [a,b]

    while len(series) < n:
        c = a + b
        series.append(c)
        a = b
        b = c

    return series

def create_graph(fibonacci_series):
    ratios = []
    for i in range(1, len(fibonacci_series)):
        if fibonacci_series[i-1] != 0:
            ratios.append(fibonacci_series[i] / fibonacci_series[i-1])
        else:
            ratios.append(0)

    plt.title('Ratio between consecutive Fibonacci numbers')
    plt.xlabel('Index') # X-axis represents the index of the ratio
    plt.ylabel('Ratio')
    plt.plot(ratios)
    plt.show()

n = int(input('Number: '))
fibonacci_series = fibo(n)
create_graph(fibonacci_series)
```
